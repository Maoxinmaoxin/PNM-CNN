# blobs方法
生成由球体随机重叠而成的结构。

```python
# blobs:生成由球体随机重叠而成的结构。
import porespy as ps

# 生成一个随机的多孔介质图像
shape = [500, 500]  # 图像的尺寸
porosity = 0.6  # 孔隙率
im = ps.generators.blobs(shape=[500, 500], 
                         porosity=0.6,
                         blobiness = 2, #控制生成斑点的平滑度和复杂度。较低的值（接近0）会生成更平滑、更圆的斑点。较高的值会生成更复杂和更不规则的斑点。
                         divs = 1 # 分辨率提升，效果不明显，设置为大于1即可提升分辨率
                         #,seed =
                         )

# 显示图像
import matplotlib.pyplot as plt
plt.imshow(im, cmap='gray')
plt.show()
```

```python
import porespy as ps
import matplotlib.pyplot as plt

# Generate images with different blobiness values
im1 = ps.generators.blobs(shape=[500, 500], porosity=0.6, blobiness=0.1, divs=2)
im2 = ps.generators.blobs(shape=[500, 500], porosity=0.6, blobiness=1, divs=2)
im3 = ps.generators.blobs(shape=[500, 500], porosity=0.6, blobiness=10, divs=2)

# Plotting the generated images
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(im1, cmap='gray')
axes[0].set_title('blobiness=0.1')
axes[0].axis('off')

axes[1].imshow(im2, cmap='gray')
axes[1].set_title('blobiness=1')
axes[1].axis('off')

axes[2].imshow(im3, cmap='gray')
axes[2].set_title('blobiness=10')
axes[2].axis('off')

plt.tight_layout()
plt.show()
```

# overlapping_spheres
通过插入随机点然后将它们膨胀到指定的半径来生成重叠的球体。

```python
import matplotlib.pyplot as plt
import numpy as np
import porespy as ps
```

```python
import inspect
print(inspect.signature(ps.generators.overlapping_spheres))
```

```python
fig, ax = plt.subplots(1, 2, figsize=[8, 4])

shape = [200, 200]
e = 0.6
r = 10
im1 = ps.generators.overlapping_spheres(shape=shape, r=r, porosity=e)
ax[0].imshow(im1)
ax[0].axis(False)
ax[0].set_title(f'radius = {r}')

r = 20
im2 = ps.generators.overlapping_spheres(shape=shape, r=r, porosity=e)
ax[1].imshow(im2)
ax[1].axis(False)
ax[1].set_title(f'radius = {r}');
```

```python
fig, ax = plt.subplots(1, 2, figsize=[8, 4])

r = 10
e = 0.8
im1 = ps.generators.overlapping_spheres(shape=shape, r=r, porosity=e)
ax[0].imshow(im1)
ax[0].axis(False)
ax[0].set_title(f'porosity = {e}')

e = 0.4
im2 = ps.generators.overlapping_spheres(shape=shape, r=r, porosity=e)
ax[1].imshow(im2)
ax[1].axis(False)
ax[1].set_title(f'porosity = {e}');
```

```python
fig, ax = plt.subplots(1, 2, figsize=[8, 4])
print('target porosity =', e)

np.random.seed(0)
mi = 1
im1 = ps.generators.overlapping_spheres(shape=shape, r=r, porosity=e, maxiter=mi)
ax[0].imshow(im1)
ax[0].axis(False)
ax[0].set_title(f'maxiter = {mi}')
print('porosity = ', ps.metrics.porosity(im1))

np.random.seed(0)
mi = 10
im2 = ps.generators.overlapping_spheres(shape=shape, r=r, porosity=e, maxiter=mi)
ax[1].imshow(im2)
ax[1].axis(False)
ax[1].set_title(f'maxiter = {mi}');
print('porosity = ', ps.metrics.porosity(im2))
```

```python
fig, ax = plt.subplots(1, 2, figsize=[8, 4])
print('target porosity =', e)

np.random.seed(0)

mi=100
im1 = ps.generators.overlapping_spheres(shape=shape, r=r, porosity=e, maxiter=mi, tol= 1e-1)
ax[0].imshow(im1)
ax[0].axis(False)
ax[0].set_title(f'tol = {1e-1}')
print('porosity = ', ps.metrics.porosity(im1))

np.random.seed(0)
im2 = ps.generators.overlapping_spheres(shape=shape, r=r, porosity=e, maxiter=mi, tol=1e-4)
ax[1].imshow(im2)
ax[1].axis(False)
ax[1].set_title(f'tol = {1e-4}');
print('porosity = ', ps.metrics.porosity(im2))
```
